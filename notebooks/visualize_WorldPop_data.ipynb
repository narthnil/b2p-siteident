```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
```

```python
raster_path = "../data/population/Rwanda_population_data_2020_UNadj_constrained.tif"
pop_raster = rasterio.open(raster_path)
pop_raster.shape
```

```python
def plot_raster(arr, title=None, figsize=(30,10)):
    """ helper function to plot a raster array """
    fig, ax = plt.subplots(figsize=figsize)
    plt.imshow(arr, cmap="Purples")
    plt.colorbar()
    # plt.title(f'Population count: {title}')
    # plt.xlabel('Column #')
    # plt.ylabel('Row #')
    plt.tick_params(
        axis='both',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        left=False,
        labelbottom=False,
    labelleft=False) # labels along the bottom edge are off
    plt.tight_layout()
    plt.savefig("worldpop_rwanda.pdf", dpi=300)
    return fig, ax
```

```python
pop_data = pop_raster.read(1)
# pop_data[pop_data < 0.] = -1.
pop_count = pop_data[pop_data>0].sum()
plot_raster(pop_data, title=pop_count)
```
