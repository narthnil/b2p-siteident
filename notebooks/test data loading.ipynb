```python
import os
import numpy as np
import matplotlib.pyplot as plt

from torch.utils.data import Dataset, Sampler, DataLoader

import sys
sys.path.append("..")

from src.data import BridgeDataset, BridgeSampler, TRAIN_METADATA
```

```python
train_metadata = dict(TRAIN_METADATA)
for country in train_metadata:
    metadata = train_metadata[country]
    for data_name in metadata:
        data = metadata[data_name]
        data["fp"] = os.path.join("..", data["fp"])
```

```python
ds = BridgeDataset(1200, use_rnd_pos=True, train_metadata=train_metadata,
                   gdf_path="../data/ground_truth/v2/")
```

```python
for _ in range(10):
    i = np.random.choice(len(ds))
    imgs, label = ds[i]
    fig, axes = plt.subplots(1, 7, figsize=(12, 20))
    
    axes[0].imshow(imgs[:, :, 1:4].astype(int))
    axes[0].set_title("OSM Image (label: {})".format(label))
    axes[1].imshow(
        imgs[:, :, 0:1].astype(int), vmin=0, vmax=100, cmap="Purples")
    axes[1].set_title("Population")
    axes[2].imshow(
        imgs[:, :, 4:5].astype(int), cmap="Reds")
    axes[2].set_title("Elevation")
    axes[3].imshow(
        imgs[:, :, 5:6].astype(int), cmap="Blues")
    axes[3].set_title("Slope")
    axes[4].imshow(
        imgs[:, :, 6:7].astype(int), cmap="Oranges")
    axes[4].set_title("Roads")
    axes[5].imshow(
        imgs[:, :, 7:8].astype(int), cmap="Blues")
    axes[5].set_title("Waterways")
    axes[6].imshow(
        imgs[:, :, 8:9].astype(int), cmap="Greys")
    axes[6].set_title("Admin bounds")
    
```
